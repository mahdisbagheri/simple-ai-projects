# **IMPORTS**

```python
from scipy.io import loadmat
import numpy as np
```

# **DATA**

## load Data

```python
mat = loadmat('Thyroid.mat')
mat
```

```python
print(mat.keys())
```

```python
data = mat['Data']
data
```

## Seperating x & y

```python
# Features
X = data[:, :7]
```

```python
# Original labels
y_original = data[:, 7]
```

## Convert Labels to binary classification

```python
# class 0 -> -1
# classes 1,2,3 -> +1
y = np.where(y_original == 0, -1, 1)
```

```python
# Given separator
w = np.array([0.048, -0.021, 0.044, 0.995, 0.000, 0.035, -0.061])
b = 0.021
```

# **LINEAR SEPARABILITY**

```python
print("=" * 60)
print("Linear Separator Evaluation")
print("=" * 60)

# Decision function
scores = X @ w + b

# Predicted labels
y_pred = np.where(scores >= 0, 1, -1)

# Margin according to the formula in the question
gamma = y * scores

# Confusion matrix components
TP = np.sum((y == 1) & (y_pred == 1))
TN = np.sum((y == -1) & (y_pred == -1))
FP = np.sum((y == -1) & (y_pred == 1))
FN = np.sum((y == 1) & (y_pred == -1))

accuracy = np.mean(y == y_pred)

# Sensitivity and Specificity
sensitivity = TP / (TP + FN) if (TP + FN) > 0 else 0
specificity = TN / (TN + FP) if (TN + FP) > 0 else 0


print(f"Number of samples : {len(y)}")
print(f"Accuracy          : {accuracy:.4f}")
print(f"Sensitivity = {sensitivity:.4f}")
print(f"Specificity = {specificity:.4f}")


print("\nConfusion Matrix")
print(f"TP = {TP}")
print(f"TN = {TN}")
print(f"FP = {FP}")
print(f"FN = {FN}")

print("\nMargin Statistics")
print(f"Minimum margin : {gamma.min():.4f}")
print(f"Average margin : {gamma.mean():.4f}")

correct = np.sum(gamma > 0)
incorrect = np.sum(gamma <= 0)

print(f"Correctly classified samples : {correct}")
print(f"Misclassified samples        : {incorrect}")

print("\nIndices of misclassified samples:")
print(np.where(gamma <= 0)[0])

print("=" * 60)
```
