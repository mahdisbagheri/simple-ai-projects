# **Clause based CSP with BT-FC**

## Version 1: Using deepcopy

```python
# Build Domains (7 valid assignments per clause)

def build_domains(clauses):
    """
    For each clause create all satisfying assignments (7 states)
    Each state is a dictionary {variable: True/False}

    We exclude the single unsatisfying combination,
    so each domain contains 7 valid states.

    Each state is represented as:
        {variable_number: True/False}
    """
    domains = []

    for clause in clauses:

        possible_states = []

        # Generate all 2^3 combinations
        for values in product([False, True], repeat=3):

            satisfied = False
            state = {}

            for lit, val in zip(clause, values):
                var = abs(lit)
                state[var] = val

                # Check literal truth
                literal_value = val if lit > 0 else not val
                if literal_value:
                    satisfied = True

            if satisfied:  # exclude all-false case
                possible_states.append(state)

        domains.append(possible_states)

    return domains
```

```python
# Build Neighbor Structure

def build_neighbors(clauses):
    """
    Two clauses are neighbors if they share a variable
    """
    n = len(clauses)
    neighbors = {i: set() for i in range(n)}

    for i in range(n):
        vars_i = set(abs(l) for l in clauses[i])
        for j in range(i + 1, n):
            vars_j = set(abs(l) for l in clauses[j])

            if vars_i & vars_j:  # intersection not empty
                neighbors[i].add(j)
                neighbors[j].add(i)

    return neighbors
```

```python
# Consistency Check Between Two States

def consistent(state1, state2):
    """
    Two states are consistent if shared variables
    have same truth value
    """
    for var in state1:
        if var in state2:
            if state1[var] != state2[var]:
                return False
    return True
```

```python
# Forward Checking with Deepcopy

def forward_check(var_index, assignment, domains, neighbors):
    
    # Create a deep copy of domains to avoid in-place modification
    new_domains = deepcopy(domains)

    for neighbor in neighbors[var_index]:

        if neighbor in assignment:
            continue

        filtered_domain = []

        for state in new_domains[neighbor]:
            if consistent(assignment[var_index], state):
                filtered_domain.append(state)

        if not filtered_domain:  # domain wiped out
            return False, None

        new_domains[neighbor] = filtered_domain

    return True, new_domains
```

```python
# MRV Heuristic

def select_unassigned_variable(assignment, domains):
    """
    Select clause with minimum remaining values (MRV)
    """
    unassigned = [i for i in range(len(domains)) if i not in assignment]

    return min(unassigned, key=lambda i: len(domains[i]))
```

```python
# Backtracking Search

def backtrack(assignment, domains, neighbors):

    if len(assignment) == len(domains):
        return assignment

    var = select_unassigned_variable(assignment, domains)

    for state in domains[var]:

        new_assignment = assignment.copy()
        new_assignment[var] = state

        ok, new_domains = forward_check(var, new_assignment, domains, neighbors)

        if ok:
            result = backtrack(new_assignment,
                               new_domains,
                               neighbors)
            if result is not None:
                return result

    return None
```

```python
# Solver Wrapper

def solve_3sat_csp(matrix):
    """
    Main solver function.

    Input:
        matrix → numpy matrix representing 3-SAT clauses

    Output:
        assignment dictionary OR None (if UNSAT)
    """
    clauses = matrix.tolist()

    domains = build_domains(clauses)
    
    neighbors = build_neighbors(clauses)

    solution = backtrack({}, domains, neighbors)

    return solution
```

```python
#  Print Final Variable Assignment

def print_final_assignment(solution, matrix):

    if solution is None:
        print("\n UNSAT (No solution found)\n")
        return

    final_values = {}

    # Merge clause states into single global assignment
    for state in solution.values():
        for var, val in state.items():
            final_values[var] = val

    num_vars = np.max(np.abs(matrix))

    print("\n SATISFYING ASSIGNMENT\n")
    print("-" * 40)

    for v in range(1, num_vars + 1):
        val = final_values.get(v, False)
        print(f"x{v:3d} = {str(val):5s}")

    print("-" * 40)

    binary = "".join(
        ['1' if final_values.get(v, False) else '0'
         for v in range(1, num_vars + 1)]
    )

    print("\nBinary representation:")
    print(binary)
    print()
```

## Vesion 2: Using Trail Stack Instead of deepcopy

```python
# Build Domains (7 satisfying assignments per clause)

def build_domains(clauses):

    domains = []

    for clause in clauses:
        possible_states = []

        for values in product([False, True], repeat=3):

            satisfied = False
            state = {}

            for lit, val in zip(clause, values):
                var = abs(lit)
                state[var] = val

                # Check if literal becomes True
                literal_value = val if lit > 0 else not val
                if literal_value:
                    satisfied = True

            # Keep only satisfying assignments
            if satisfied:
                possible_states.append(state)

        domains.append(possible_states)

    return domains
```

```python
# Build Constraint Graph (Neighbors)

def build_neighbors(clauses):

    n = len(clauses)
    neighbors = {i: set() for i in range(n)}

    for i in range(n):
        vars_i = set(abs(l) for l in clauses[i])

        for j in range(i + 1, n):
            vars_j = set(abs(l) for l in clauses[j])

            # If intersection not empty → constraint exists
            if vars_i & vars_j:
                neighbors[i].add(j)
                neighbors[j].add(i)

    return neighbors
```

```python
# Consistency Check Between Two Clause Assignments

def consistent(state1, state2):

    for var in state1:
        if var in state2:
            if state1[var] != state2[var]:
                return False
    return True
```

```python
# Forward Checking (No Deepcopy)

def forward_check(var_idx, assignment, domains, neighbors, trail):
    """
    Perform Forward Checking after assigning var_idx.

    Instead of copying domains, we:
    - Modify domains in place
    - Record all domain changes in 'trail'
      so they can be undone if needed

    If any neighbor domain becomes empty → return False.
    """

    for nb in neighbors[var_idx]:

        if nb in assignment:
            continue

        old_domain = domains[nb]
        new_domain = []

        # Keep only consistent states
        for state in old_domain:
            if consistent(assignment[var_idx], state):
                new_domain.append(state)

        # Domain wipe-out → failure
        if len(new_domain) == 0:
            return False

        # If domain changed, record it for undo
        if new_domain != old_domain:
            trail.append((nb, old_domain))
            domains[nb] = new_domain

    return True
```

```python
# Undo Domain Changes (Rollback)

def undo(trail, domains):
    """
    Restore all domains modified during
    the last forward checking step.
    """
    for var, old_domain in reversed(trail):
        domains[var] = old_domain

    trail.clear()
```

```python
# MRV (Minimum Remaining Values Heuristic)

def select_next_variable(assignment, domains):

    unassigned = [i for i in range(len(domains))
                  if i not in assignment]

    return min(unassigned, key=lambda i: len(domains[i]))
```

```python
# Backtracking Search (No Deepcopy)

def backtrack(assignment, domains, neighbors):
    """
    Recursive Backtracking with:
        - Forward Checking
        - MRV heuristic
        - Trail-based domain restoration
    """

    # If all clauses assigned → solution found
    if len(assignment) == len(domains):
        return assignment

    var = select_next_variable(assignment, domains)

    for state in domains[var]:

        # Assign current state
        assignment[var] = state

        trail = []

        # Apply forward checking
        if forward_check(var, assignment, domains, neighbors, trail):

            result = backtrack(assignment, domains, neighbors)

            if result is not None:
                return result

        # Undo changes (rollback)
        undo(trail, domains)
        del assignment[var]

    return None
```

```python
# Solver Wrapper

def solve_3sat_csp(matrix):

    clauses = matrix.tolist()

    domains = build_domains(clauses)

    neighbors = build_neighbors(clauses)

    solution = backtrack({}, domains, neighbors)

    return solution
```

## Final Vesion: Time & Space Complexity Check + Version 2

```python
# GLOBAL COUNTERS

COUNTERS = {
    "BACKTRACK_CALLS": 0,
    "CONSISTENCY_CHECKS": 0,
    "PRUNED_STATES": 0,
    "MAX_RECURSION_DEPTH": 0,
    "MAX_TOTAL_DOMAIN_SIZE": 0
}
```

```python
def update_domain_space(domains):
    total = sum(len(d) for d in domains)
    COUNTERS["MAX_TOTAL_DOMAIN_SIZE"] = max(COUNTERS["MAX_TOTAL_DOMAIN_SIZE"], total)
```

```python
# Build Domains (7 satisfying assignments per clause)

def build_domains(clauses):
    domains = []

    for clause in clauses:
        possible_states = []

        for values in product([False, True], repeat=3):

            satisfied = False
            state = {}

            for lit, val in zip(clause, values):
                var = abs(lit)
                state[var] = val

                literal_value = val if lit > 0 else not val
                if literal_value:
                    satisfied = True

            if satisfied:
                possible_states.append(state)

        domains.append(possible_states)

    update_domain_space(domains)
    return domains
```

```python
# Build Neighbors

def build_neighbors(clauses):
    n = len(clauses)
    neighbors = {i: set() for i in range(n)}

    for i in range(n):
        vars_i = set(abs(l) for l in clauses[i])

        for j in range(i + 1, n):
            vars_j = set(abs(l) for l in clauses[j])

            if vars_i & vars_j:
                neighbors[i].add(j)
                neighbors[j].add(i)

    return neighbors
```

```python
# Check consistency

def consistent(state1, state2):
    for var in state1:
        if var in state2:
            COUNTERS["CONSISTENCY_CHECKS"] += 1
            if state1[var] != state2[var]:
                return False
    return True
```

```python
# Forward Checking

def forward_check(var_idx, assignment, domains, neighbors, trail):
    for nb in neighbors[var_idx]:

        if nb in assignment:
            continue

        old_domain = domains[nb]
        new_domain = []

        for state in old_domain:
            if consistent(assignment[var_idx], state):
                new_domain.append(state)
            else:
                COUNTERS["PRUNED_STATES"] += 1

        if not new_domain:
            return False

        if new_domain != old_domain:
            trail.append((nb, old_domain))
            domains[nb] = new_domain
            update_domain_space(domains)

    return True
```

```python
# Undo changes

def undo(trail, domains):
    for var, old_domain in reversed(trail):
        domains[var] = old_domain
    trail.clear()
```

```python
# MRV

def select_next_variable(assignment, domains):
    unassigned = [i for i in range(len(domains)) if i not in assignment]
    return min(unassigned, key=lambda i: len(domains[i]))
```

```python
# Backtracking with counters

def backtrack(assignment, domains, neighbors, depth=0):

    COUNTERS["BACKTRACK_CALLS"] += 1
    COUNTERS["MAX_RECURSION_DEPTH"] = max(COUNTERS["MAX_RECURSION_DEPTH"], depth)

    if len(assignment) == len(domains):
        return assignment

    var = select_next_variable(assignment, domains)

    for state in domains[var]:

        assignment[var] = state
        trail = []

        if forward_check(var, assignment, domains, neighbors, trail):
            result = backtrack(assignment, domains, neighbors, depth + 1)
            if result is not None:
                return result

        undo(trail, domains)
        del assignment[var]

    return None
```

```python
# Solver wrapper

def solve_3sat_csp(matrix):

    # Reset counters each run
    for k in COUNTERS:
        COUNTERS[k] = 0

    clauses = matrix.tolist()
    domains = build_domains(clauses)
    neighbors = build_neighbors(clauses)

    solution = backtrack({}, domains, neighbors)

    return solution, COUNTERS
```

```python
def extract_variable_assignment(solution):
    final = {}
    for clause_state in solution.values():
        final.update(clause_state)
    return final
```

```python
def run_instance(name, matrix):
    print("=" * 50)
    print(f"Solving instance: {name}")
    print("=" * 50)

    start = time.perf_counter()

    solution, stats = solve_3sat_csp(matrix)

    end = time.perf_counter()
    runtime = end - start

    if solution is None:
        print("Result: UNSAT (No solution found)")
    else:
        print("Result: SAT (Solution found)")

        var_solution = extract_variable_assignment(solution)

        print("\nSolution (Variable Assignment):")
        for v in sorted(var_solution):
            print(f"x{v} = {var_solution[v]}")

    print("\n--- Complexity Metrics ---")
    print(f"{'Backtrack Calls(Time Complexity):':25s} {stats['BACKTRACK_CALLS']}")
    print(f"{'Consistency Checks:':25s} {stats['CONSISTENCY_CHECKS']}")
    print(f"{'Pruned States (FC):':25s} {stats['PRUNED_STATES']}")
    print(f"{'Max Recursion Depth(Space Complexit):':25s} {stats['MAX_RECURSION_DEPTH']}")
    print(f"{'Max Domain Size (total):':25s} {stats['MAX_TOTAL_DOMAIN_SIZE']}")
    print(f"{'Runtime (seconds):':25s} {runtime:.6f}")

    print("\n")
```

# **Walksat**

```python
import random
import time
```

```python
def evaluate_clause(clause, assignment):
    for lit in clause:
        var = abs(lit) - 1
        if (lit > 0 and assignment[var]) or (lit < 0 and not assignment[var]):
            return True
    return False
```

```python
def num_unsatisfied(clauses, assignment):
    count = 0
    for c in clauses:
        if not evaluate_clause(c, assignment):
            count += 1
    return count
```

```python
def flip(var, assignment):
    assignment[var] = not assignment[var]
```

```python
def walksat(clauses, n_vars, max_tries, max_flips, p):
    start_time = time.perf_counter()

    for _ in range(max_tries):

        assignment = [random.choice([True, False]) for _ in range(n_vars)]

        for _ in range(max_flips):

            unsat_clauses = [c for c in clauses if not evaluate_clause(c, assignment)]

            if not unsat_clauses:
                end_time = time.perf_counter()
                return assignment, (end_time - start_time)

            clause = random.choice(unsat_clauses)

            if random.random() < p:
                lit = random.choice(clause)
                var = abs(lit) - 1
                flip(var, assignment)

            else:
                best_var = None
                best_score = float("inf")

                for lit in clause:
                    var = abs(lit) - 1
                    flip(var, assignment)
                    score = num_unsatisfied(clauses, assignment)
                    flip(var, assignment)

                    if score < best_score:
                        best_score = score
                        best_var = var

                if best_var is None:
                    lit = random.choice(clause)
                    best_var = abs(lit) - 1

                flip(best_var, assignment)

    end_time = time.perf_counter()
    return None, (end_time - start_time)
```
