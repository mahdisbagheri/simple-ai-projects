# **NAIVE BAYES**

## Train Bernoulli Naive Bayes

```python
def train_naive_bayes(X_train, y_train):

    # Prior probabilities
    p_democrat = np.mean(y_train == 1)
    p_republican = np.mean(y_train == 0)

    # Laplace smoothing
    alpha = (
        np.sum(X_train[y_train == 1], axis=0) + 1
    ) / (
        np.sum(y_train == 1) + 2
    )

    beta = (
        np.sum(X_train[y_train == 0], axis=0) + 1
    ) / (
        np.sum(y_train == 0) + 2
    )

    return p_democrat, p_republican, alpha, beta
```

## Prediction

```python
def predict_naive_bayes(X_test,
                        p_democrat,
                        p_republican,
                        alpha,
                        beta):

    predictions = []

    for x in X_test:

        log_democrat = np.log(p_democrat)
        log_republican = np.log(p_republican)

        log_democrat += np.sum(
            x * np.log(alpha)
            + (1 - x) * np.log(1 - alpha)
        )

        log_republican += np.sum(
            x * np.log(beta)
            + (1 - x) * np.log(1 - beta)
        )

        if log_democrat >= log_republican:
            predictions.append(1)
        else:
            predictions.append(0)

    return np.array(predictions)
```

# **EVALUATION METRICS**

```python
def confusion_matrix_binary(y_true, y_pred):

    TP = np.sum((y_true == 1) & (y_pred == 1))
    TN = np.sum((y_true == 0) & (y_pred == 0))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))

    return TP, TN, FP, FN
```

# **CROSS VALIDATION**

## 5-Fold Cross Validation

```python
fold_size = 65

errors = []
accuracies = []
precisions = []
recalls = []

print("=" * 80)
print("5-FOLD CROSS VALIDATION RESULTS")
print("=" * 80)

for fold in range(5):

    start = fold * fold_size
    end = start + fold_size

    X_test = X[start:end]
    y_test = y[start:end]

    X_train = np.concatenate(
        [X[:start], X[end:]],
        axis=0
    )

    y_train = np.concatenate(
        [y[:start], y[end:]]
    )

    p_democrat, p_republican, alpha, beta = \
        train_naive_bayes(X_train, y_train)

    y_pred = predict_naive_bayes(
        X_test,
        p_democrat,
        p_republican,
        alpha,
        beta
    )

    TP, TN, FP, FN = \
        confusion_matrix_binary(y_test, y_pred)

    accuracy = (TP + TN) / len(y_test)
    error = 1 - accuracy

    precision = TP / (TP + FP) \
        if (TP + FP) > 0 else 0

    recall = TP / (TP + FN) \
        if (TP + FN) > 0 else 0

    accuracies.append(accuracy)
    errors.append(error)
    precisions.append(precision)
    recalls.append(recall)

    print("\n")
    print("-" * 80)
    print(f"FOLD {fold+1}")
    print("-" * 80)

    print(f"Training Samples : {len(X_train)}")
    print(f"Testing Samples  : {len(X_test)}")

    print("\nConfusion Matrix")
    print(f"TP = {TP}")
    print(f"TN = {TN}")
    print(f"FP = {FP}")
    print(f"FN = {FN}")

    print("\nPerformance")
    print(f"Accuracy  = {accuracy*100:.2f}%")
    print(f"Error     = {error*100:.2f}%")
    print(f"Precision = {precision*100:.2f}%")
    print(f"Recall    = {recall*100:.2f}%")
```

## Overall Cross Validation Result

```python
print("\n")
print("=" * 80)
print("OVERALL CROSS VALIDATION RESULT")
print("=" * 80)

for i in range(5):

    print(
        f"Fold {i+1}: "
        f"Accuracy={accuracies[i]*100:.2f}% , "
        f"Error={errors[i]*100:.2f}%"
    )

print("\nAverage Results")

print(f"Average Accuracy  = {np.mean(accuracies)*100:.2f}%")
print(f"Average Error     = {np.mean(errors)*100:.2f}%")
print(f"Average Precision = {np.mean(precisions)*100:.2f}%")
print(f"Average Recall    = {np.mean(recalls)*100:.2f}%")
```

# **FEATURE ANALYSIS**

## Train On Full Dataset For Feature Analysis

```python
p_democrat, p_republican, alpha, beta = \
    train_naive_bayes(X, y)

# Weight formula from lecture slides

weights = (
    np.log(alpha / (1 - alpha))
    -
    np.log(beta / (1 - beta))
)

tendency = []

for w in weights:

    if w > 0:
        tendency.append("Democrat")

    elif w < 0:
        tendency.append("Republican")

    else:
        tendency.append("Neutral")
```

## Feature Analysis Table

```python
feature_table = pd.DataFrame({

    "Feature": feature_names,

    "Alpha=P(x=1|Democrat)": alpha,

    "Beta=P(x=1|Republican)": beta,

    "Weight": weights,

    "AbsWeight": np.abs(weights),

    "Tendency": tendency

})
```
